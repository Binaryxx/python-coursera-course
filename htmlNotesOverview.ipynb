Key words enclosed in angled brackets are called tags  
Tags communicate how to display the content  

##### Tags:  
<\!DOCTYPE html>
> Indicates the document is an html doc  

<\html>  
> root element of the html page  

<\title>  
> title of the document; shows in the browser's title bar  
  
<\head>  
> contains meta info about the page  
  
<\body>  
> what is displayed in the page  
  
<\a>  
> anchor tag  
  
Indentation determines the hierarchy of the tags  
> characteristics are inherited by the children
